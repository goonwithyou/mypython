- 截图

需要设定一个含有四个元素的元组box，(left, upper, right, lower),(100, 100, 400, 400)截取的是从100，100开始的300x300的图片。
```python
box = (100, 100, 400, 400)
region = im.crop(box)

region = region.transpose(Image.ROTATE_180)
# 把处理过的region图，粘贴到im图中的box位置
im.paste(region, box)
```

- Drawing postscript

```python
from PIL import PSDraw

im = Image.open('C:\\Users\\Cap\\Pictures\\2.jpeg')
title = 'This is my draw'
box = (100, 100, 500, 500)
ps = PSDraw.PSDraw()
ps.begin_document(title)
ps.image(box, im, 75)
ps.rectangle(box)
ps.setfont('HelveticaNarrow-Bold', 36)
ps.text((200, 200), title)

ps.end_document()
```
