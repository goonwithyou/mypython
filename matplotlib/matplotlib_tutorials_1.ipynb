# pyplot

## General Concepts
---
1. **Figure**. The figure keeps track of all the child **Axes**, a smattering of special **artists**(titles, figure legends, etc), and the **canvas**.
    - a figure can have any number of **Axes**, but to be useful shoule have at least one.
2. **Axes**. this is what you think of as a plot, it is the region of the image with data space.
    - a given figure can contain many **Axes** object, but a given **Axes** object can only be in one **Figure**.
    - the Axes contains two (or three) **Axis** objects which take care of the data limits.
    - each Axes has a **title**, an **x-label**, and a **y-label**.
3. **Axis**. the take care of setting the **graph limit** and **generating the ticks** and **ticklabels**.
    - the location of ticks is determined by a **Locator** object and the ticklabel string are formatted by a **Formatter**.
4. **artist**. basically everything you can see on the figure is an artist(even the **Figure, Axes, Axis objects**).
    - this include **text** object, **Line2D** objects, **collection** objects, **Patch** objects.
    - when the figure is rendered, all of the artists are drawn to the **canvas**.
    - Most **artists** are tied to an **Axes**. an artist cannot be shared by multiple Axes.

## Logarithmic and other nonlinear axes
---

```python
from matplotlib.ticker import NullFormatter


np.random.seed(19680801)

y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure(1)

plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

plt.subplot(223)
plt.plot(x, y - np.mean(y))
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.subplots_adjust(top=0.92, bottom=0.08, 
                    left=0.10, right = 0.95, hspace=0.4,
                    wspace=0.35)

plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

def currency(x, pos):
    """The two args are the value and tick position"""
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s

formatter = FuncFormatter(currency)

fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(range(len(group_names)), group_data, tick_label=group_names)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
      title='Company Revenue')
ax.xaxis.set_major_formatter(formatter)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(range(len(group_names)), group_data, tick_label=group_names)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

ax.axvline(group_mean, ls='--', color='r')

for group in [3, 5, 8]:
    ax.text(14500, group, 'New Company', 
            fontsize=10, verticalalignment='center')

ax.title.set(y=1.05)

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
      title='Company Revenue')
ax.xaxis.set_major_formatter(formatter)
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
fig.subplots_adjust(right=.8)

plt.show()
```

## 3.2.1 Styling with cycler

```python
from cycler import cycler
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
yy.shape
```

```python
default_cycler = cycler('color', ['r', 'g', 'b', 'y']) + \
                    cycler('linestyle', ['-', '--', ':', '-.'])

custom_cycler = cycler('color', ['c', 'm', 'y', 'k']) + \
                    cycler('lw', [1, 2, 3, 4])
    
    
plt.rc('lines', linewidth=4)
plt.rc('axes', prop_cycle=default_cycler)

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(yy)
ax1.set_title('set default color cycle to rgby')

ax2.set_prop_cycle(custom_cycler)
ax2.plot(yy)
ax2.set_title('set axes color cycle to cmyk')

fig.subplots_adjust(hspace=0.5)
plt.show()
```
